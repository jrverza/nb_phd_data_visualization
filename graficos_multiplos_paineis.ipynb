```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import matplotlib as mpl
import seaborn as sns
%matplotlib inline 

#muda o tamanho da fonte universalmente
mpl.rcParams['font.size']=14
```

```python
df = pd.read_csv('data_for_exercises/spectroscopy.csv')
```

```python
df.head()  #comando para ver as 5 primeiras linhas do df
```

```python
#para ver quantas linhas e quantas colunas meu df possui
df.shape 
```

```python
#minuindo a densidade dos dados, isso pode ser errado, mas para esse projeto é bom
df = df.iloc[::10, : ]
```

```python
#atribuindo os dados às variáveis 
x90 = df['X 90 Kohm'].dropna()
y90 = df['Y 90 Kohm'].dropna()
x90_clear = df['X 90 Kohm-clear'].dropna()
y90_clear = df['Y 90 Kohm-clear'].dropna()

x60 = df['X 60 Kohm'].dropna()
y60 = df['Y 60 Kohm'].dropna()
x60_clear = df['X 60 Kohm-clear'].dropna()
y60_clear = df['Y 60 Kohm-clear'].dropna()

x30 = df['X 30 Kohm'].dropna()
y30 = df['Y 30 Kohm'].dropna()
x30_clear = df['X 30 Kohm-clear'].dropna()
y30_clear = df['Y 30 Kohm-clear'].dropna()

#preparando o plot em multiplos painéis

fig = plt.figure(1,figsize=(5,5))
gs = gridspec.GridSpec(4,4) #estamos criando uma area de plotagem 4x4
gs.update(wspace=0.2, hspace=0.25) #espaos entre os paineis, vertical e horizontal

#gerando o primeiro painel(plot)
xtr_subplot = fig.add_subplot(gs[0:4, 0:2]) #quer dizer que estamos usando todas as 4 linhas, porém 2 colunas

#plot data for left panel
plt.plot(x90, y90, linestyle='none', marker='^', label='90', mfc='w', markersize=8)
plt.plot(x60, y60, linestyle='none', marker='o', label='60', mfc='w', markersize=8)
plt.plot(x30, y30, linestyle='none', marker='s', label='30', mfc='w', markersize=8)

#fazendo os ticks manualmente
xticks = np.arange(300,701,200)
yticks = np.arange(60,121,20)

#informações dos ticks
plt.minorticks_on()
plt.tick_params(direction='in',which='minor', length=5, bottom=True, top=True, left=True, right=True)
plt.tick_params(direction='in',which='major', length=10, bottom=True, top=True, left=True, right=True)
plt.xticks(xticks)
plt.yticks(yticks)

#limits do plot
plt.xlim(300,801)
plt.ylim(59,121)

#legenda e eixos
plt.legend()
plt.ylabel('Transparency %')

plt.text(645,52,'Wavelength (nm)') #escreve começando no x=645 e y=52


#gerando secundo painel
xtr_subsplot = fig.add_subplot(gs[0:2,2:4])

plt.plot(x90_clear, y90_clear, linestyle='none', marker='^', label='90 clear', mfc='w', markersize=8)
plt.plot(x60_clear, y60_clear, linestyle='none', marker='o', label='60 clear', mfc='w', markersize=8)
plt.plot(x30_clear, y30_clear, linestyle='none', marker='s', label='30 clear', mfc='w', markersize=8)

#fazendo os ticks manualmente
xticks = np.arange(100,801,200)
yticks = np.arange(60,121,20)

#informações dos ticks
plt.minorticks_on()
plt.tick_params(direction='in',which='minor', length=5, bottom=True, top=True, left=True, right=True)
plt.tick_params(direction='in',which='major', length=10, bottom=True, top=True, left=True, right=True)
plt.tick_params(labelbottom=False, labeltop=False, labelright=True, labelleft=False)
plt.xticks(xticks)
plt.yticks(yticks)

#limits do plot
plt.xlim(300,801)
plt.ylim(59,121)

#legenda e eixos
plt.legend(fontsize='x-small')



#gerando terceiro painel
xtr_subsplot = fig.add_subplot(gs[2:4,2:4])
plt.plot(x90_clear, y90_clear, linestyle='none', marker='^', label='90 clear', mfc='w', markersize=8)
plt.plot(x60_clear, y60_clear, linestyle='none', marker='o', label='60 clear', mfc='w', markersize=8)
plt.plot(x30_clear, y30_clear, linestyle='none', marker='s', label='30 clear', mfc='w', markersize=8)


#fazendo os ticks manualmente
xticks = np.arange(100,801,200)
yticks = np.arange(60,121,20)


#informações dos ticks
plt.minorticks_on()
plt.tick_params(direction='in',which='minor', length=5, bottom=True, top=True, left=True, right=True)
plt.tick_params(direction='in',which='major', length=10, bottom=True, top=True, left=True, right=True)
plt.tick_params(labelbottom=True, labeltop=False, labelright=True, labelleft=False)
plt.xticks(xticks)
plt.yticks(yticks)


#limits do plot
plt.xlim(300,801)
plt.ylim(59,121)


plt.legend(fontsize='x-small')

#Export figure
#plt.savefig('data_for_exercises/plotting/multi_panel_error.png', dpi=300,bbox_inches="tight")
```
